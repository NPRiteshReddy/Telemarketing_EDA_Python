#### Data Cleaning

```python
# Droping NAs

bank_marketing=bank_marketing.dropna()
bank_marketing.isnull().sum()
```

### Feature Engineering

```python
# Converting columns ( day and month ) into new column as Date ( eg: 2017-05-05 ) as datetime format

bank_marketing['day'] = bank_marketing['day'].astype(str)+ ' ' + bank_marketing['month']
bank_marketing['Date'] = pd.to_datetime(bank_marketing['month'], format='mixed')
```

```python
bank_marketing['Date'] = pd.to_datetime(bank_marketing['day'], format='mixed')
```

```python
# Spliting Jobedu column into two as Job_type and Education

bank_marketing[["jobedu","education"]]=bank_marketing['jobedu'].str.split(',', expand=True)
```

```python
bank_marketing[["Duration","units"]]=bank_marketing['duration'].str.split(' ', expand=True)
```

```python
bank_marketing['Duration'] = bank_marketing['Duration'].astype(float)
```

```python
bank_marketing["job_type"]=bank_marketing["jobedu"]
```

```python
bank_marketing['Duration_minutes'] = bank_marketing['Duration'] / 60
```

```python
# Dropping the unwanted columns

bank_marketing.drop(columns=['month','jobedu','day','duration','Duration','units'], inplace=True)
```

```python
Q1 = count["age"].quantile(0.25)
Q3 = count["age"].quantile(0.75)

# Calculate the IQR
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the data to remove outliers
no_outliers_age = count[(count["age"] >= lower_bound) & (count["age"] <= upper_bound)]
no_outliers_age
```

```python
no_outliers_age.shape
```

```python
Q1 = count["campaign"].quantile(0.25)
Q3 = count["campaign"].quantile(0.75)

# Calculate the IQR
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the data to remove outliers
no_outliers_campaign = count[(count["campaign"] >= lower_bound) & (count["campaign"] <= upper_bound)]
no_outliers_campaign
```

```python
Q1 = count["pdays"].quantile(0.25)
Q3 = count["pdays"].quantile(0.75)

# Calculate the IQR
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the data to remove outliers
no_outliers_pdays = count[(count["pdays"] >= lower_bound) & (count["pdays"] <= upper_bound)]
no_outliers_pdays
```

```python
Q1 = count["Duration_minutes"].quantile(0.25)
Q3 = count["Duration_minutes"].quantile(0.75)

# Calculate the IQR
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the data to remove outliers
no_outliers_Duration_minutes = count[(count["Duration_minutes"] >= lower_bound) & (count["Duration_minutes"] <= upper_bound)]
no_outliers_Duration_minutes
```

## RESPONSE ANALYSIS

```python
plt.figure(figsize=(15,10))

plt.subplot(2,2,1)
sns.countplot(y="poutcome",data=bank_marketing,hue="response")

plt.subplot(2,2,3)
sns.countplot(x="previous",data=bank_marketing,hue="response")

plt.subplot(2,2,4)
sns.countplot(x="campaign",data=bank_marketing,hue="response")

plt.subplot(2,2,2)
bank_marketing["response"].value_counts().plot.pie(ylabel="Response",autopct="%.2f%%")

plt.show()
```

```python
response_counts = bank_marketing.groupby(['contact', 'response']).size().unstack(fill_value=0)

response_counts.plot(kind='bar', stacked=True, figsize=(5, 3))
plt.xlabel('contact')
plt.ylabel('Response_Count')
plt.title('Stacked Bar Chart of Responses by contact')
plt.legend(title='Response')
plt.show()
```

#### KEY TAKEAWAYS
- Overall response is very bad (11.7% )
- Poutcome based on response analysis tells that the coustumers not reached out numbers are very high and very low success rate in it.
- Number of times the customer has been reached in the previous campaigns or for any of the other products is near to zero so is the respone relating to it.
- Campaign chart shows the low response ratio compaired to number of campaign in which customer is contacted.

#### RESPONSE TREND ANALYSIS

```python
bank_marketing['Date'] = pd.to_datetime(bank_marketing['Date'])

response_counts = bank_marketing.groupby(['Date', 'response']).size().unstack(fill_value=0)

response_counts.plot(kind='line', marker='o', figsize=(20, 6))
plt.xlabel('Date')
plt.ylabel('Count')
plt.title('Trends of "Yes" and "No" Responses over Time')
plt.legend(title='Response')
plt.grid(True)
plt.show()
```

```python
bank_marketing['Date'] = pd.to_datetime(bank_marketing['Date'])

targeted_counts = bank_marketing.groupby(['Date','targeted']).size().unstack(fill_value=0)

targeted_counts.plot(kind='line', marker='o', figsize=(20, 6))
plt.xlabel('Date')
plt.ylabel('Count')
plt.title('Trends of "Yes" and "No" Targeted_counts over Time')
plt.legend(title='Targeted')
plt.grid(True)
plt.show()
```

#### KEY TAKEAWAYS
- Response trend as show indicates high spikes of "No" response in between the months of May - July.
- It was only on 5th of May where there was "Yes" response.
- Though the targeted analysis show good amount of trageted trend but cloudnt leverage it on the response.
