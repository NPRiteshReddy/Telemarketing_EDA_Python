#### Data Cleaning
- There are few NA's which needs to be cleared.
- Need to identify the different data types while doing.
- Replacing NA's with mean for continuous data and manual or mode for discrete data.

```python
bm["age"].fillna(bm["age"].mean(),inplace=True)
```

```python
bm["month"].value_counts()
```

```python
bm["month"].fillna("may, 2017",inplace=True)
```
