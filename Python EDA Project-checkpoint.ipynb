```python
Q1 = bm["pdays"].quantile(0.25)
Q3 = bm["pdays"].quantile(0.75)

# Calculate the IQR
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the data to remove outliers
df_no_outliers_pdays = bm[(bm["pdays"] >= lower_bound) & (bm["pdays"] <= upper_bound)]
```

```python
df_no_outliers_pdays.shape
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df_no_outliers_pdays["pdays"])
plt.title("Box Plot of Data without Outliers")
plt.show()
```

```python
campaign_Q1 = bm["campaign"].quantile(0.25)
campaign_Q3 = bm["campaign"].quantile(0.75)

# Calculate the IQR
campaign_IQR = campaign_Q3 - campaign_Q1

# Define the lower and upper bounds for outliers
campaign_lower_bound = campaign_Q1 - 1.5 * campaign_IQR
campaign_upper_bound = campaign_Q3 + 1.5 * campaign_IQR

# Filter the data to remove outliers
df_no_outliers_campaign = bm[(bm["campaign"] >= campaign_lower_bound) & (bm["campaign"] <= campaign_upper_bound)]
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df_no_outliers_campaign["campaign"])
plt.title("Box Plot of Data without Outliers")
plt.show()
```
